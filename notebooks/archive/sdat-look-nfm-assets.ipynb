```python
import pandas as pd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy

%matplotlib inline
```

```python
df = pd.read_excel('data/Shipston Assets 8.10 Cambridge snapshot (first 200 assets).xlsx', header=1)
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())

plt.plot(df['x'], df['y'], 'x')

ax.gridlines(draw_labels=True)
```

```python
dict_1 = {}

for i in range(len(df['Asset Type'])):
    dict_1[df['Asset Type'][i]] = df['Asset Type'][i]

    
for key in dict_1:
    dict_1[key] = 0
    for i in range(len(df['Asset Type'])):
        if key == df['Asset Type'][i]:
            dict_1[key] += 1

asset_list = []
num_asset_list = []

for key in dict_1:
    asset_list.append(key)
    num_asset_list.append(dict_1[key])
    
df_types = pd.DataFrame({'number': num_asset_list},
                        index=asset_list)

df_types.plot.pie(y='number', figsize=(5, 5))
```
