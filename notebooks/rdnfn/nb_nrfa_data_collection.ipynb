```python
import nrfa_api

collector = nrfa_api.NrfaDataCollector()
collector.available_stations
cover_data_source = 'lcm2000' #land cover map 2000
print(collector.get_static_data(1001,[cover_data_source]))

import pandas as pd
gauge_ids = pd.read_csv('CAMELS_GB_climatic_attributes.csv')['gauge_id'].values

cover_df = pd.DataFrame()

for gauge_id in gauge_ids[:]:
    cover_data = collector.get_static_data(gauge_id,[cover_data_source])
    cover_df = cover_df.append(cover_data['data'])

cover_df = cover_df.rename(columns = {'id':'gauge_id'})
```

```python
cover_df = cover_df.fillna(0)
cover_df.to_csv("NRFA_landcover_attributes_{}.csv".format(cover_data_source), index=False)
```
