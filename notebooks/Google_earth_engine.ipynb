```python
import ee
ee.Authenticate()
```

```python
ee.Initialize()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dem =  ee.Image('USGS/SRTMGL1_003')
#max= 544 for 0.04 degree span
resolution  = 100
latitude = 52.0607
longitude = -1.6228
mat = np.zeros((resolution,resolution))

for i, x_val in enumerate(np.linspace(longitude-0.02, longitude+0.02, resolution)):
    for j, y_val in enumerate(np.linspace(latitude-0.02, latitude+0.02, resolution)):
        xy = ee.Geometry.Point([x_val, y_val])
        elev = dem.sample(xy, 30).first().get('elevation').getInfo()
        if i%5==0 and j%5==0:
            print(i, j, elev)
        mat[i,j] = elev
        
plt.imshow(mat)
```
